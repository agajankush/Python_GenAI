### Analyzing traffic Data with Datetime

```python
df = pd.read_csv('website_traffic_data_datetime.csv')
df
```

```python
df.dtypes
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.set_index('Date', inplace=True)
```

```python
df
```

```python
monthly_visits = df["Visits"].resample('ME').sum()
```

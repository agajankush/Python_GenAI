## Introduction to Pandas: Series and DataFrames

```python
import pandas as pd
```

```python
data = {
    "Category": ["Electronics", "Electronics", "Furniture", "Furniture", "Apparel"],
    "Product": ['Laptop', 'Mouse', 'Chair', 'Table', 'Shoes'],
    'Sales': [1200, 100, 400, 300, 50]
}
df = pd.DataFrame(data)
```

```python
df
```

```python
grouped_data = df.groupby(["Category"]).sum()
```

```python
grouped_data
```

```python
# Groupby multiple columns
grouped_data = df.groupby(["Category", "Product"])["Sales"].sum()
grouped_data
```

```python
# Multiple Aggregate function on a single dataframe
grouped_data = df.groupby("Category")["Sales"].agg(['sum', 'mean', 'median'])
```

```python
grouped_data
```

### Time Series Data Manipulations in Pandas

```python
data = {
    'OrderID': [1001, 1002, 1003, 1004],
    'OrderDate': ['2023-11-01', '2023-11-02', '2023-11-03', '2023-11-04'],
    'Sales': [250, 400, 150, 300]
}
df = pd.DataFrame(data)
```

```python
df
```

```python
df.dtypes
```

```python
df['OrderDate'] = pd.to_datetime(df['OrderDate'])
```

```python
df.dtypes
```

```python
df.set_index('OrderDate', inplace=True)
df
```

```python
df['Year'] = df.index.year
df['Month'] = df.index.month
df['Day'] = df.index.day
```

```python
df
```

```python
# Resample is used to group data by a specific time frequency
# W -> weekly
weekly_sales = df['Sales'].resample('W').sum()
weekly_sales
```
